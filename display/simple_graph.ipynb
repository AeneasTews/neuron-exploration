```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
```

```python
pos = {1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}  # explicitly set positions
#pos = nx.spring_layout(G)  automatically create positions

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edge_color": "black",
    "linewidths": 5,
    "width": 5,
}
```

```python
nx.draw_networkx(G, pos, **options)

# margins
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
# -----------------------------------Directed Graph----------------------------------
G = nx.DiGraph([(0, 3), (1, 3), (2, 4), (2, 1), (3, 5), (3, 6), (4, 6), (5, 6)])
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)
```
